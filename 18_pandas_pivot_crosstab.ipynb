### Pivot

```python
df = pd.DataFrame({'foo': ['one', 'one', 'one', 'two', 'two',
                           'two'],
                   'bar': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'baz': [1, 2, 3, 4, 5, 6],
                   'zoo': ['x', 'y', 'z', 'q', 'w', 't']})
df
```

```python
df.pivot(index='foo', columns='bar', values='baz')
```

```python
df.pivot(index='foo', columns='bar')
```

```python
df = pd.DataFrame({
       "lev1": [1, 1, 1, 2, 2, 2],
       "lev2": [1, 1, 2, 1, 1, 2],
       "lev3": [1, 2, 1, 2, 1, 2],
       "lev4": [1, 2, 3, 4, 5, 6],
       "values": [0, 1, 2, 3, 4, 5]})
df
```

```python
df.pivot(index='lev1', columns=['lev2', 'lev3'], values='values')
```

```python
df.pivot(index=['lev1', 'lev2'], columns=['lev3'], values='values')
```

### Pivot table

```python
df = pd.DataFrame({"A": ["foo", "foo", "foo", "foo", "foo",
                         "bar", "bar", "bar", "bar"],
                   "B": ["one", "one", "one", "two", "two",
                         "one", "one", "two", "two"],
                   "C": ["small", "large", "large", "small",
                         "small", "large", "small", "small",
                         "large"],
                   "D": [1, 2, 2, 3, 3, 4, 5, 6, 7],
                   "E": [2, 4, 5, 5, 6, 6, 8, 9, 9]})
df
```

```python
table = pd.pivot_table(df, values='D', index=['A','B'], 
                       columns=['C'], aggfunc=np.sum)
table
```

```python
table = pd.pivot_table(df, values=['D', 'E'], index=['A', 'C'],
                    aggfunc={'D': np.mean,
                             'E': [min, max, np.mean]})
table
```

### Sort both index and columns

```python
import numpy as np
import pandas as pd

df = pd.DataFrame({

"TIME":["FQ1","FQ2","FQ2","FQ2"],
"NAME":["Robert",'Miranda',"Robert","Robert"],
"TOTAL":[900,42,360,2000],
"TYPE":["Air","Ground","Air","Ground"],
"GROUP":["A","A","A","A"]})

df
```

```python
pt = pd.pivot_table(data=df,
               values =["TOTAL"], aggfunc = (np.sum),
               index = ["GROUP","TYPE","NAME"],
               columns = "TIME",
               fill_value=0,
               margins = True)
pt
```

```python
pt.sort_values(['GROUP','TYPE',('TOTAL','All')], 
               ascending=[True, True, False])
```
